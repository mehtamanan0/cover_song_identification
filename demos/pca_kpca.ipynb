```python
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
from plotly import tools
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_circles
```

```python
#original
np.random.seed(0)
X, y = make_circles(n_samples=400, factor=.3, noise=.05)
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
groupedo = df.groupby('label')
for key, group in groupedo:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
pca = PCA()
X_pca = pca.fit_transform(X)
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
pca_df = DataFrame(dict(x=X_pca[:,0], y=X_pca[:,1], label=y))
groupedp = pca_df.groupby('label')
for key, group in groupedp:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
kpca = KernelPCA(kernel="rbf", fit_inverse_transform=True, gamma=10)
X_kpca = kpca.fit_transform(X)
kpca_df = DataFrame(dict(x=X_kpca[:,0], y=X_kpca[:,1], label=y))
groupedk = kpca_df.groupby('label')
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
for key, group in groupedk:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```
